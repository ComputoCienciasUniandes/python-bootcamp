## But what is a random walk?

Imagine that a walker start a walk at point $O$, he gave steps of lenght $l\ $ in any direction (random angle $\theta$) in the x-y plane.
What is the probability of findind the walker at a distance $d > l$ from the point $O$ at time $t\ $?



We can study this problem by making use of statistics theory in that case we have to solve the equation:

Or we can simulate this that is the aim of this course. 

# 2D-Random Walk 

```python
%pylab inline
```

```python
l = 1 # Length of my step
n_steps = 100 # lenght of my walk


walk_x = np.zeros(n_steps)
walk_y = np.zeros(n_steps)

for i in range(n_steps-1):
    theta = np.random.random()*2*np.pi
    walk_x[i+1] = walk_x[i] + l*cos(theta)
    walk_y[i+1] = walk_y[i] + l*sin(theta)
```

## Let's see how this looks

```python
plot(walk_x, walk_y)
scatter(walk_x[0], walk_y[0], c='r', s=80)
scatter(walk_x[-1], walk_y[-1], s=80)
```

What can we get from this simulation?

Lets fine the total walk distance $d$

```python
print "d = ", np.sqrt(walk_x[-1]**2 + walk_y[-1]**2)
```
